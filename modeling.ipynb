```python
import pandas as pd
import os
normal, imablance6, imablance10, imablance15, imablance20, imablance25, imablance30, imablance35 = [],[],[],[],[],[],[],[]
dirnormal = 'normal'
dir_imbalance6 = 'imbalance/6g'
dir_imbalance10 = 'imbalance/10g'
dir_imbalance15 = 'imbalance/15g'
dir_imbalance20 = 'imbalance/20g'
dir_imbalance25 = 'imbalance/25g'
dir_imbalance30 = 'imbalance/30g'
dir_imbalance35 = 'imbalance/35g'

directories = [dirnormal,dir_imbalance6, dir_imbalance10, dir_imbalance15, dir_imbalance20, dir_imbalance25, dir_imbalance30, dir_imbalance35]
paths = [normal, imablance6, imablance10, imablance15, imablance20, imablance25, imablance30, imablance35]
path = 0
for dir in directories:
  for file in os.listdir(dir):
    if file.endswith(".csv"):
      paths[path].append(directories[path] +"/" +file)
  print(len(paths[path]))
  print(paths[path])
  path+=1
```

```python
# This loops through the lists that was created in the above code and puts all of the data from the csv files into a single data frame
# data frames are still separated by normal and the levels of imbalance
column_names = ["Tachometer signal",
                "Underhang axial direction",
                "Underhand radial direction",
                "Underhang tangential direction",
                "Overhang axial direction",
                "Overhang radial direction",
                "Overhang tangential direction",
                "Microphone"]

df_normal, df_imablance6, df_imablance10, df_imablance15, df_imablance20, df_imablance25, df_imablance30, df_imablance35 = pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame()
# df = pd.read_csv(paths[0][0], names=column_names, header=None)
# df_temp = pd.read_csv(paths[0][1], names=column_names, header=None)

# df = df.append(df_temp)

for x in paths[0]:
    df_temp = pd.read_csv(x, names=column_names, header=None)
    df_normal = df_normal.append(df_temp)
for x in paths[1]:
    df_temp = pd.read_csv(x, names=column_names, header=None)
    df_imablance6 = df_imablance6.append(df_temp)
for x in paths[2]:
    df_temp = pd.read_csv(x, names=column_names, header=None)
    df_imablance10 = df_imablance10.append(df_temp)
for x in paths[3]:
    df_temp = pd.read_csv(x, names=column_names, header=None)
    df_imablance15 = df_imablance15.append(df_temp)
for x in paths[4]:
    df_temp = pd.read_csv(x, names=column_names, header=None)
    df_imablance20 = df_imablance20.append(df_temp)
for x in paths[5]:
    df_temp = pd.read_csv(x, names=column_names, header=None)
    df_imablance25 = df_imablance25.append(df_temp)
for x in paths[6]:
    df_temp = pd.read_csv(x, names=column_names, header=None)
    df_imablance30 = df_imablance30.append(df_temp)
for x in paths[7]:
    df_temp = pd.read_csv(x, names=column_names, header=None)
    df_imablance35 = df_imablance35.append(df_temp)
```

```python
# here we add a new column to the data frames which will be our label
labels = ["normal", "6g", "10g", "15g", "20g", "25g", "30g", "35g"]
dfs = [df_normal, df_imablance6, df_imablance10, df_imablance15, df_imablance20, df_imablance25,df_imablance30,df_imablance35]
pos = 0
for i in range(len(dfs)):
  label = []
  for i in range(len(dfs[pos])):
    label.append(labels[pos])
  dfs[pos]['label'] = label
  pos+=1
```

```python
# We will create another dataframe that is made up of random samples from all the other data frames
# This will allow us to visualize the data without crashing google colab
visualize_df = pd.DataFrame()
for df in dfs:
  temp_df = df.sample(frac=0.20, replace=True, random_state=1)
  visualize_df = visualize_df.append(temp_df, ignore_index= True)

visualize_df.describe()
```

```python
# visulaize the data as per report 1 requirements
import matplotlib.pyplot as plt
import seaborn
seaborn.set()

f = plt.figure()
f.set_figwidth(8)
f.set_figheight(8)

N, bins, patches = plt.hist(visualize_df['label'],rwidth=0.7, bins=range(1,9), align='left')

patches[0].set_facecolor('red')
patches[1].set_facecolor('orange')
patches[3].set_facecolor('yellow')
patches[4].set_facecolor('green')
patches[5].set_facecolor('blue')
patches[6].set_facecolor('purple')

plt.title('Histogram visualization of the data classes, based on labels created')
plt.ylabel('Count')
plt.xlabel("Data Classes")


```
